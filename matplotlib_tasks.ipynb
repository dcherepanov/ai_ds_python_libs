Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.

Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
fig, ax = plt.subplots(1)
ax.set_title(r'График $f(t)$')
ax.set_xlabel(r'Значения $t$')
ax.set_ylabel(r'Значения $f$')
ax.plot(t, f, color='green')
ax.set_xlim(0.5, 9.5)
ax.set_ylim(-2.5, 2.5)
plt.show()
```

В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте $y_1$,в правом верхнем - $y_2$, в левом нижнем - $y_3$, в правом нижнем - $y_4$.
```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
fig, ax = plt.subplots(2, 2, figsize=(8, 6))
ax[0][0].set_xlim(-5, 5)
ax[0][0].set_title(r'График $y_1$')
ax[0][0].plot(x, y1)
ax[0][1].set_title(r'График $y_2$')
ax[0][1].plot(x, y2)
ax[1][0].set_title(r'График $y_3$')
ax[1][0].plot(x, y3)
ax[1][1].set_title(r'График $y_4$')
ax[1][1].plot(x, y4)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```
